```python
import matplotlib.pyplot as plt
import numpy as np
X=2*np.random.randn(100,1)
y=5+X**2+2*np.random.randn(100,1)
plt.plot(X,y,"g.")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_features=PolynomialFeatures(degree=2,include_bias=False)
X_poly=poly_features.fit_transform(X)
print(X[0])
print(X_poly[0])
```

```python
lin_reg=LinearRegression()
lin_reg.fit(X_poly,y)
lin_reg.intercept_,lin_reg.coef_
```

# <font color="red">NOTE</font> the coefficient array is showing two elements, the first is $\sim$ 0 and the second is $\sim$ 1 which is closely reflecting our model to generate the data so in our case, taking into account the definition: $y=ax^2+bx+c$ 

```python
print ("c:",np.round(lin_reg.intercept_))
print ("a:",np.round(lin_reg.coef_[0][1]))
print ("b:",np.round(lin_reg.coef_[0][0]))
```

## Note how the model in RED fits the data set in BLUE

```python
plt.plot(X,y,"b.",label="dataset")
X2=np.linspace(-6,6,1000)
plt.plot(X2,lin_reg.coef_[0][1]*X2**2+5,"r-", label="model")
plt.legend()
plt.show()
```
