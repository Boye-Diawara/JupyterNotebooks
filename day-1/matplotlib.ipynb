```

def squarewave(x):
    return (1 if (x>0) else 0)
print (squarewave(1))
```

```

def sq(x):
    return(x**2)
```

```
from math import exp
def sigma(x):
    return (1/(1+exp(-x)))
```

```
import matplotlib.pyplot as plt

points=100
xmin=-10
xmax=10
xlist=list(map(lambda x:xmin+float(xmax-xmin)*x/points,range(points+1))) 
ylist=list(map(squarewave,xlist))
plt.plot(list(xlist),list(ylist))
plt.ylabel('Square Wave plot')
plt.grid(True, which='both')
plt.show()
```

```
zlist=list(map(sq,xlist))
plt.plot(xlist,zlist)
plt.ylabel('Parabola plot')
plt.grid(True, which='both')
#plt.display()
plt.show()
```

```

klist=list(map(sigma,xlist))
plt.plot(xlist,klist)
plt.ylabel('Sigmoid plot')
plt.grid(True, which='both')
plt.show()


```
