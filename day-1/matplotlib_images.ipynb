```python
def mand(c):
    z=0
    for _ in range(100):
        z=z*z+c
        if (abs(z)>2): break
    return _

X=np.linspace(-1,1,100)
Y=np.linspace(-1,1,100)
img=[]
for y in Y:
    for x in X:
        img.append(mand(complex(x,y)))
image=np.array(img)
image.shape=(100,100)
plt.imshow(image)
```
